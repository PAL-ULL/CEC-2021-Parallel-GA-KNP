```python
from os import listdir
import pandas as pd

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
files = [('../data/rankings/50', 'Instances N = 50'), 
         ('../data/rankings/100', 'Instances N = 100'), 
         ('../data/rankings/500', 'Instances N = 500'), 
         ('../data/rankings/1000', 'Instances N = 1000')]
```

```python
for path, title in files:
    print(f'\n\n\t{title}')
    #print('=' * 125)
    for file in listdir(path):
        #print(f'{file[:file.find("_")]}')
        df = pd.read_csv(f'{path}/{file}', sep='&', engine='python')
        df.index.names = [file[:file.find("_")]]
        print(df.iloc[[0, 5, 10, 14, 15], [0, 1, 3, 4, 6, 7, 8, 9]].to_latex())
        #print('=' * 125)
```
